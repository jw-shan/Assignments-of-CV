```python
import matplotlib.pyplot as plt 
import matplotlib.image as img
import numpy as np
# from PIL import Image
```

```python
pic = img.imread('test12.png') 
pic.shape 

plt.imshow(pic) # 显示图片
plt.axis('off') # 不显示坐标轴
plt.show()
```

```python
rows,cols,_ = pic.shape
k = 8
```

```python
for row in range(0,rows,k):
    for col in range(0,cols,k):
        for channel in range(4):
            ave = np.mean(pic[row:row+k, col:col+k, channel])
            pic[row:row+k, col:col+k, channel] = np.ones([k,k],dtype=np.float32)*ave
            
```

```python
plt.imshow(pic) # 显示图片
plt.axis('off') # 不显示坐标轴
plt.show()
```
