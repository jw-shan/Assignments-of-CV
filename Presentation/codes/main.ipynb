```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
```

## Pixelization

```python
pic = img.imread('test6.png')
pic_pix = pic.copy()
```

```python
plt.imshow(pic) 
plt.axis('off') 
plt.show()
```

```python
print(pic.shape[:2])
```

```python
rows,cols,_ = pic.shape
k = 8
for row in range(0,rows,k):
    for col in range(0,cols,k):
        for channel in range(4):
            ave = np.mean(pic[row:row+k, col:col+k, channel])
            pic_pix[row:row+k, col:col+k, channel] = np.ones([k,k],dtype=np.float32)*ave
            
```

```python
plt.imshow(pic_pix) 
plt.axis('off') 
plt.show()
```
